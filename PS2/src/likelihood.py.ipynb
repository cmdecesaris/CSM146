```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
```

```python
fig, ax = plt.subplots(1,1)
num_zeros=4
n=10
store=[]
store1=[]
store2=[]
store3=[]
x =np.arange(0,1,0.01)
for p in x:
    
    likelihood = bernoulli.pmf(1, p)**6 * bernoulli.pmf(0,p)**4
    likelihood1 = bernoulli.pmf(1, p)**3 * bernoulli.pmf(0,p)**2
    likelihood2 = bernoulli.pmf(1, p)**60 * bernoulli.pmf(0,p)**40
    likelihood3 = bernoulli.pmf(1, p)**5 * bernoulli.pmf(0,p)**5

    # print(likelihood)
    store.append(likelihood)    
    store1.append(likelihood1)
    store2.append(likelihood2)
    store3.append(likelihood3)

 
plt.plot(0.6, max(store), "o", c='r')  
plt.plot(x,store)
plt.xticks(np.arange(0,1,0.1))
plt.style.use("seaborn-darkgrid")
plt.xlabel(r"$\theta$")
plt.ylabel(r"L($\theta$)")
plt.title("Likelihood Graph N=5")


plt.show()

plt.figure(1)
 
plt.plot(0.6, max(store2), "o", c='r')  
plt.plot(x,store2)
plt.xticks(np.arange(0,1,0.1))
plt.style.use("seaborn-darkgrid")
plt.xlabel(r"$\theta$")
plt.ylabel(r"L($\theta$)")
plt.title("Likelihood Graph N=100")

plt.figure(2)
fig, ax = plt.subplots(1,1)
num_zeros=4
n=10
store=[]
store1=[]
store2=[]
store3=[]
x =np.arange(0,1,0.01)
for p in x:
    
    likelihood = bernoulli.pmf(1, p)**6 * bernoulli.pmf(0,p)**4
    likelihood1 = bernoulli.pmf(1, p)**3 * bernoulli.pmf(0,p)**2
    likelihood2 = bernoulli.pmf(1, p)**60 * bernoulli.pmf(0,p)**40
    likelihood3 = bernoulli.pmf(1, p)**5 * bernoulli.pmf(0,p)**5

    # print(likelihood)
    store.append(likelihood)    
    store1.append(likelihood1)
    store2.append(likelihood2)
    store3.append(likelihood3)

 
plt.plot(0.6, max(store), "o", c='r')  
plt.plot(x,store)
plt.xticks(np.arange(0,1,0.1))
plt.style.use("seaborn-darkgrid")
plt.xlabel(r"$\theta$")
plt.ylabel(r"L($\theta$)")
plt.title("Likelihood Graph N=5")


plt.show()

plt.figure(1)
 
plt.plot(0.5, max(store3), "o", c='r')  
plt.plot(x,store3)
plt.xticks(np.arange(0,1,0.1))
plt.style.use("seaborn-darkgrid")
plt.xlabel(r"$\theta$")
plt.ylabel(r"L($\theta$)")
plt.title("Likelihood Graph N=10")
```
